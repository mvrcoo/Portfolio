```python
import pandas as pd

# Replace 'data.csv' with the correct file path
file_path = "data.csv"

try:
    # Load the CSV file
    data = pd.read_csv(file_path)
    print("Data loaded successfully!")

    # Display all rows by adjusting Pandas display settings
    pd.set_option('display.max_rows', None)
    display(data)

    # Check if 'Adjusted Close' column exists
    if 'Adjusted Close' in data.columns:
        adj_close = data['Adjusted Close']
        print("Adjusted Close column extracted:")
        print(adj_close)
    else:
        print("The 'Adjusted Close' column does not exist in the data.")

except FileNotFoundError:
    print(f"File not found: {file_path}")
except pd.errors.EmptyDataError:
    print("The file is empty!")
except Exception as e:
    print(f"An error occurred: {e}")
```
